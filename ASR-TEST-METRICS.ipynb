# ASR Metrics

```python
asr = df.columns[4:]
```

```python
wers, cers = dict(), dict()
```

```python
def add_and_print(df, row_name, cols=asr):
    labels = werpy.normalize(df.loc[:, 'label'].values)

    local_wers, local_cers = list(), list()

    for col in asr:

        predictions = werpy.normalize(df.loc[:, col].values)
        wer = werpy.wer(labels, predictions)
        cer = char_error_rate(labels, predictions).item()

        local_wers.append(wer)
        local_cers.append(cer)

        print(f'Распознавание с {col}, значение wer: {wer:.2f}, значениие cer: {cer:.2f}')

    wers[row_name] = local_wers
    cers[row_name] = local_cers
```

Для всех аудиозаписей

```python
add_and_print(df, 'Все записи')
```

Для слогов

```python
df_s = df[(df['type'] == 0)]
add_and_print(df_s, 'Слоги')
```

Для слов

```python
df_w = df[(df['type'] == 1)]
add_and_print(df_w, 'Слова')
```

Для аудиозаписей с ошибками

```python
df_e = df[(df['error'] == 1)]
add_and_print(df_e, 'Речь с ошибками')
```

Для аудиозаписей без ошибок

```python
df_c = df[(df['error'] == 0)]
add_and_print(df_c, 'Речь без ошибок')
```

Слоги без ошибок

```python
df_es = df[(df['error'] == 0) & (df['type'] == 0)]
add_and_print(df_es, 'Слоги без ошибок')
```

Слова без ошибок

```python
df_ew = df[(df['error'] == 0) & (df['type'] == 1)]
add_and_print(df_ew, 'Слова без ошибок')
```

Слоги с ошибками

```python
df_cs = df[(df['error'] == 1) & (df['type'] == 0)]
add_and_print(df_cs, 'Слоги с ошибками')
```

Слова с ошибками

```python
df_ew = df[(df['error'] == 1) & (df['type'] == 1)]
add_and_print(df_cs, 'Слова с ошибками')
```
